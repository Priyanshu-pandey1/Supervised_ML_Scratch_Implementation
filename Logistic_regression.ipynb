```python
import numpy as np

class LogisticRegression:
    # 1. Fixed n_uter -> n_iter
    def __init__(self, learning_rate=0.1, n_iter=1000):
        # 2. Removed the trailing commas to avoid creating tuples
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter
        
    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
        
    def fit(self, X, y):
        m, n = X.shape
        self.bias = 0
        self.weights = np.zeros(n)

        for i in range(self.n_iter):
            z = self.bias + np.dot(X, self.weights)
            y_pred = self._sigmoid(z)

            # 3. Added np.sum for the bias gradient
            db = (1/m) * np.sum(y_pred - y)
            dw = (1/m) * np.dot(X.T, (y_pred - y))

            self.bias -= self.lr * db
            self.weights -= self.lr * dw
            
    def get_probabilities(self, X):
        z = self.bias + np.dot(X, self.weights)
        return self._sigmoid(z)
        
    def predict(self, X, threshold=0.5):
        probabilities = self.get_probabilities(X)
        y_pred_bool = probabilities >= threshold
        return y_pred_bool.astype(int)
```
